```
n = 1
alls = 0
while True:
  x = input(f'enter {n} value of x:-- ')
  x = float(x)
  print(x, end = ':--')
  total = x**2*10/2
  print(total)
  print('\n')
  alls = alls + total
  print(f'Total sum is {alls}')
  print('------'*30)
  n += 1
```
